```python
import pandas as pd

# Example DataFrame
data = {
    'element_1': [1, 2, 3, 5, 5, 5],
    'element_2': [4, 4, 4, 3, 2, 1],
    'globalsim': [0.8, 0.6, 0.7, 0.9, 0.85, 0.8]
}

dfessai = pd.DataFrame(data)

# Lists
ids = [1, 2, 3]
targets = [4, 5]

# Function to calculate average similarity
def calculate_average_similarity(ids, targets, df):
    results = []

    for id in ids:
        similarities = []

        for target in targets:
            # Find similarities for both (id, target) and (target, id)
            sim6 = dfessai[(dfessai['element_1'] == id) & (dfessai['element_2'] == target)]['globalsim']
            sim7 = dfessai[(dfessai['element_1'] == target) & (dfessai['element_2'] == id)]['globalsim']
            
            # Collect all found similarities
            if not sim6.empty:
                similarities.extend(sim6.tolist())
            if not sim7.empty:
                similarities.extend(sim7.tolist())
        
        # Calculate the average similarity
        if similarities:
            avg_similarity = sum(similarities) / len(similarities)
        else:
            avg_similarity = None  # Or some default value if no similarities are found
        
        # Store the result
        results.append({
            'id': id,
            'average_similarity': avg_similarity
        })

    return results

# Calculate average similarities
average_similarities = calculate_average_similarity(ids, targets, df)

# Convert results to DataFrame
results_df = pd.DataFrame(average_similarities)
print(results_df)
```
